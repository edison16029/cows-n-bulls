```python
import json
import requests
```

```python
def load_words(filename):
    with open(filename,'r') as f:
        return json.load(f)
```

```python
combined_words = load_words("./_CNBFourLetterWordsTWDef.json")
# combined_words = ['myth','spam','ssde','hugi']
```

```python
def retrieveTWDifficulty(word,endpoint="definition/"):
    url = "https://twinword-word-graph-dictionary.p.rapidapi.com/" + endpoint
    querystring = {"entry":word}
    headers = {
        'x-rapidapi-host': "twinword-word-graph-dictionary.p.rapidapi.com",
        'x-rapidapi-key': "cc7dc146fbmshfed28857fb4ea89p170cb0jsnd674115956bc"
        }
    response = requests.request("GET", url, headers=headers, params=querystring)
    return response.json()
```

```python
# Find words which has a value for definition endpoint 
defined_words = []
undefined_words = []

count = 0
defined_count = 0 
undefined_count = 0
for _,val in combined_words.items():
    word = val['word']
    response = retrieveTWDifficulty(word)
#     print(response)
    if 'meaning' in response.keys() :
#         print(word," is defined, Definition : ",response['meaning'])
        defined_words.append(word)
        defined_count+=1
    else:
#         print("Error for ",word," : ",response['result_msg'])
        undefined_words.append(word)
        undefined_count+=1
    
    count+=1
    if count%100 == 0:
        print("Count : ",count)
        print("Defined : ",defined_count)
        print(defined_words)
        print("Undefined :",undefined_count)
        print(undefined_words)
        print("===========================================================================")

print("Count : ",count)
print("Defined : ",defined_count)
print("Undefined :",undefined_count)



```

```python
# Find words which has a value for difficulty endpoint 
defined_words = {}
undefined_words = {}

count = 0
defined_count = 0 
undefined_count = 0
for word in combined_words:
    temp = {}
    temp['word'] = word
    response = retrieveTWDifficulty(word,"difficulty/")
    if response['result_code'] == '200':
#         print(word," is defined with difficulty : ",response['ten_degree'])
        temp['difficulty'] = response['ten_degree']
        defined_words[defined_count] = temp
        defined_count+=1
    else:
#         print("Error for ",word," : ",response['result_msg'])
        temp['error'] = response['result_msg']
        undefined_words[undefined_count] = temp
        undefined_count+=1
    
    count+=1
    if count%100 == 0:
        print("Count : ",count)
        print("Defined : ",defined_count)
        print("Undefined :",undefined_count)
        print("===========================================================================")

print("Count : ",count)
print("Defined : ",defined_count)
print("Undefined :",undefined_count)

with open('defined.json','w') as f:
    json.dump(defined_words,f,indent=4)
print("File 1 Written")

with open('undefined.json','w') as f:
    json.dump(undefined_words,f,indent=4)
print("File 2 Written")

```
