```python
import json
```

```python
def load_words(filename):
    with open(filename,'r') as f:
        return json.load(f)
```

```python
filename = '../analysis/_CNBFourLetterWordsScored.json'
```

```python
scored_words = load_words(filename)
```

```python
len(scored_words)
```

```python
from collections import Counter
diff_counter = Counter(item['difficulty'] for item in scored_words)
for c in range(1,11):
    print(c," : ",diff_counter[str(c)])
```

```python
for item in scored_words:
    print(item)
```

```python

word_list = []
for item in scored_words:
    word_dict = {}
    word_dict['word'] = item['word']
    word_dict['defined_in'] = item['defined_in']
    word_dict['1gram-score'] = item['1gram-score']
    word_dict['is_profane'] = item['is_profane']
    word_dict['level'] = item['level']
    word_dict['is_wc'] = 'true'
    if item['frequency'] == 'null':
        fq = 2.5
    elif item['frequency'] == '-1': 
        fq = 2
    else:
        fq = float(item['frequency'])
    word_dict['frequency'] = fq
              
    if item['difficulty'] == 'null':
        word_dict['difficulty'] = 4
    else:
        word_dict['difficulty'] = int(item['difficulty'])
    word_list.append(word_dict)
#     print(word_dict)
#     break
print("Count : ",len(word_list))
        
# word_list
```

```python
#Make difficulty > 7 as wp words
wc_words = 0
for item in word_list:
    if item['difficulty'] >=8:
        item['is_wc'] = 'false'
    else:
        wc_words +=1
```

```python
count = 0
for item in word_list:
    if item['frequency'] < 3 and item['is_wc']=='true':
        item['is_wc'] = 'false'
#         print(item)
        count +=1
print(count)
```

```python
with open('../analysis/_CNBFourLetterWordsFinal.json','w') as f:
    json.dump(word_list,f,indent=1)
```
